```python
# To remove the row where uf='DADO NÃO CADASTRADO' in columns ult_evento, nome, subs, and uso
mining_clean=mining_clean.drop(mining_clean[mining_clean['ult_evento'] == 'DADO NÃO CADASTRADO'].index)
mining_clean=mining_clean.drop(mining_clean[mining_clean['nome'] == 'DADO NÃO CADASTRADO'].index)
mining_clean=mining_clean.drop(mining_clean[mining_clean['subs'] == 'DADO NÃO CADASTRADO'].index)
mining_clean=mining_clean.drop(mining_clean[mining_clean['uso'] == 'DADO NÃO CADASTRADO'].index)
print('Shape of mining_clean: ',mining_clean.shape)
print('shape of mining: ',mining.shape)

print('ult_evento: ',len(mining_clean.loc[mining_clean['ult_evento']=='DADO NÃO CADASTRADO'])) # 4 rows
print('nome: ',len(mining_clean.loc[mining_clean['nome']=='DADO NÃO CADASTRADO'])) # 7 rows
print('subs: ',len(mining_clean.loc[mining_clean['subs']=='DADO NÃO CADASTRADO'])) # 9 rows
print('uso: ',len(mining_clean.loc[mining_clean['uso']=='DADO NÃO CADASTRADO'])) # 9 rows

#Number of unique values in each column after cleaning of mininng data
mining_clean.nunique(axis=0) #now 340 types of minerals
```

```python
mining_clean.nunique(axis=0)
```

```python
#When the id are same, all column values, except objectid and globalid, are same
#So I do the following to remove the duplicate rows
id_mc=mining_clean['id'].unique() #similar to above except for the whole data
similar_no_max=0 #to find the max number of times where id values are same
for xl,x in enumerate(id_mc):
  ind = mining_clean[mining_clean['id']==x].index.tolist()
  yl=len(ind)
  if yl > 1:
    #print('----------',xl,x,'---------')
    #print('No. of rows with same id value: ',yl)
    chk1=mining_clean.loc[ind,:]
    #print(chk1)
    #print(ind)
    #To drop the columns objectid and globalid.. these are the columns which differ when id are same
    chk1=chk1.drop(['objectid','globalid'],axis=1)
    chk1_all=chk1['id']+'_'+chk1['numero'].map(str)+'_'+chk1['ano'].map(str)+'_'+chk1['area_ha'].map(str)+'_'+chk1['fase']+'_'+chk1['ult_evento']+'_'+chk1['nome']+'_'+chk1['subs']+'_'+chk1['uso']+'_'+chk1['uf']+'_'+chk1['status']+'_'+chk1['shape_Length'].map(str)+'_'+chk1['shape_Area'].map(str)
    chk1_uni=chk1_all.unique()
    #print('chk1_uni:',chk1_uni)
    chk1_uni_len=len(chk1_uni)
    if chk1_uni_len == 1:
      ind2=ind[1:]
      mining_clean=mining_clean.drop(ind2)
    #------------------------------------------------------
    #if chk1_uni_len is not equal to 1 then all columns (except objectid and globalid) are not identical
    if chk1_uni_len > 1: 
      print('----------',xl,x,'---------')
      print('No. of rows with same id value: ',yl)
      print(chk1_uni)
      print('--------------------------------------------')
  if similar_no_max < yl:
    similar_no_max=yl
print('similar_no_max: ',similar_no_max)
#time taken for above process: 38m 9s
```
